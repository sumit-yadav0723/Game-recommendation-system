## Feature Extraction from Text data (Method defination)

```python
import re
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer
from string import punctuation, digits
from collections import Counter
from wordcloud import WordCloud
from pprint import pprint 
from sklearn.feature_extraction.text import TfidfVectorizer
import gensim
import gensim.corpora as corpora
from gensim.models import CoherenceModel
```

```python
lemmatizer = WordNetLemmatizer()
stop = set(stopwords.words('english'))
stop = ENGLISH_STOP_WORDS.union(stop) # combine stop words from different sources
game_word = stop.union(['game','play','free','best','fun','new','enjoy','1',
                   'real','epic','challenge','app','make','mobile','win', 
                   'android','100','2020','2021','1000','version', 
                   'youtube','facebook','instagram','twitter','multiplayer','singleplayer'
                   "xbox","live","gameplay","experience","player","mode","feature","character",
                   "world","help","live","edition","episode","fantastic",
                   "action","adventure","shooter","roleplaying","miscellaneous","strategy","arcade",
                   "thirdperson","firstperson","3d","2d","sports","scifi","fantasy","racing","simulation",
                   "traditional","realtime","openworld","survival","music","horror",
                   'nintendo64', 'wii', 'gamecube', 'switch', 'xbox', 'dreamcast','playstation3',
                    'playstation', 'playstation2', 'pc','gameboyadvance', 'playstation4', '3ds', 'xbox360',
                    'wiiu','playstationvita', 'ds', 'xboxone', 'psp', 'playstation5','xboxseriesx', 'stadia']) # add common game words
```

```python
# Transform Roman no. to integer
def remove_romanno(text):
    ''' text -> input string 
    output -> string with roman no. transformed to integers'''

    roman={"i":'1',
    "ii":'2',
    "iii":'3',
    "iv":'4',
    "v":'5',
    "vi":'6',
    "vii":'7',
    "viii":'8',
    "ix":'9',
    "x":'10'}
    return [roman[word] if word in roman.keys() else word for word in text ]

# function for removing url from text
def remove_url(text):
    '''Remove URL from a string'''

    text = re.sub(r'''(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'".,<>?«»“”‘’]))''', 
                  '', text)
    return " ".join(re.sub("([^0-9A-Za-z \t])|(\w+:\/\/\S+)", "", text).split())

# Remove Email from a String
def remove_email(text):
    '''Remove E-Mail from string'''

    return re.sub('\S*@\S*\s?', '', text)

def remove_brackets(text):
    '''Remove Bracket from string'''

    text=re.sub(r"\[([A-Za-z0-9_]+)\]",'',text)
    text=re.sub(r"\(([A-Za-z0-9_]+)\)",'',text)
    text=re.sub(r"\{([A-Za-z0-9_]+)\}",'',text)
    return text

# Function for expanding most common contractions
def decontraction(phrase):
    '''Expanding most common contractions in string'''
    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase

# To group words which have collective meaning
def contraction(phrase):
    '''Merge pair of words which have combined meaning'''
    phrase=re.sub(r"single player","singleplayer",phrase)
    phrase=re.sub(r"multi player","multiplayer",phrase)
    phrase=re.sub(r"electronic art","electronicart",phrase)
    phrase=re.sub(r"open world","openworld",phrase)
    phrase=re.sub(r"role playing","roleplaying",phrase)
    phrase=re.sub(r"world war","worldwar",phrase)
    return phrase

# remove punctuations
def remove_punct(text):
    '''remove punctuations'''
    table=str.maketrans('','',punctuation)
    return text.translate(table)

# Tokenize words and remove stopwords
def remove_stopwords(text):
    '''Tokenize words and remove stopwords'''
    return [word for word in word_tokenize(text) if word.lower() not in stop]

def lemmatization(text):
    '''lemmetize words'''
    return [lemmatizer.lemmatize(word.lower(), pos='n') for word in text]

def remove_gamewords(text):
    '''Remove common Game terms'''
    return [word for word in text if word not in game_word]

# remove only digits that are not joined to char (eg. 3d is not removed)
def remove_digits(text):
    '''remove only digits that are not joined to char'''
    return [word for word in text if not word.isdigit()]

def keep_alpha(text):
    '''keep words which contains alphabet only'''
    return [word for word in text if word.isalpha()]

def remove_escape_sequences(text):
    '''remove Escape Sequences(\n,\t,\r)'''
    text=text.replace(r"\n","")
    text=text.replace(r"\t","")
    text=text.replace(r"\r","")
    return text
```

### Cleaning description and name 

**Preprocess and clean game description**
- Before doing that, the game descriptions need to be cleaned and preprocessed first. The following steps were performed:

    - Remove escape sequences. An example is \n which is used to break sentences
    - Remove emails such as gamename@gmail.com, usually developer email address
    - Remove URLs, which were included in game descriptions to link to other pages
    - Perform decontraction, which expands words like "won't" to "will not"
    - Remove punctuations
    - Remove stopwords (eg. to, with, and) and general game and app words (eg. game, play, fun)
    - Perform lemmatization, which reduces words to their base form, such as "playing" to "play", or "adventurous" to "adventure"
    - Remove digits such as "2021", "1 v 1", "100%" but not digits in words like '3d'
    - Description is not unique -> preprocess individual plot before merging eg. 8710

```python
# clean description
def transform_des(text):
    text=text.strip().lower()
    text=remove_url(text)
    text=remove_email(text)
    text=decontraction(text)
    text=remove_stopwords(text)
    text=lemmatization(text)
    text=remove_romanno(text)
    text=remove_gamewords(text)
    text=remove_digits(text)
    text=keep_alpha(text)
    return text

# clean game name
def proper_name(name):
    name=name.strip(" ',-;").lower()
    name=decontraction(name)
    name=name.split(":")[0]
    name=remove_punct(name)
    name=remove_brackets(name)
    name=lemmatization(remove_stopwords(name))
    name=remove_romanno(name)
    name=remove_gamewords(name)
    name=keep_alpha(name)
    # return name
    text=''
    for s in name:
        text+=s+" "
    
    return text.strip()
```

```python
# Ngrams
from nltk.util import ngrams

def generate_ngrams(text, n):
    words = word_tokenize(text)
    return [' '.join(ngram) for ngram in list(get_data(ngrams(words, n))) if not all(w in stop for w in ngram)] # exclude if all are stopwords


# in newer versions of python, raising StopIteration exception to end a generator, which is used in ngram, is deprecated
def get_data(gen):
    try:
        for elem in gen:
            yield elem
    except (RuntimeError, StopIteration):
        return
```

```python
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=3):
    """
    Compute c_v coherence for various number of topics

    Parameters:
    ----------
    dictionary : Gensim dictionary
    corpus : Gensim corpus
    texts : List of input texts
    limit : Max num of topics

    Returns:
    -------
    model_list : List of LDA topic models
    coherence_values : Coherence values corresponding to the LDA model with respective number of topics
    """ 
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        model = gensim.models.ldamodel.LdaModel(corpus=corpus, num_topics=num_topics, id2word=id2word, random_state=100, 
                                                update_every=1, chunksize=100, passes=10, alpha='auto', per_word_topics=True)                                        
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

    return model_list, coherence_values
# start,limit,step=2,10,1
# model_list, coherence_values = compute_coherence_values(dictionary=id2word, corpus=corpus, texts=description_words, start=start, limit=limit, step=step)
```
