```python
import pandas as pd
import numpy as np
game_large = pd.read_csv('all_games.csv')
game_des = pd.read_csv('games_of_all_time.csv')
```

```python
import requests
from bs4 import BeautifulSoup
headers={'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/34.0.1847.131 Safari/537.36'}
```

```python
l=[]
for i in game_des['url']:
    if(len(l)%500==0):
        print(len(l))
    web_page=requests.get(i,headers=headers).text
    soup=BeautifulSoup(web_page)
    l.append(str(soup.find('img',class_='product_image large_image')).split()[-1].split('"')[1])
    break
```

```python
img_url=[]
```

```python
l=[]
for i in game_des['url']:
    if(len(l)%1000==0):
        print(len(l))
    try:
        web_page=requests.get(i,headers=headers).text
        soup=BeautifulSoup(web_page)
        l.append(str(soup.find('img',class_='product_image large_image')).split()[-1].split('"')[1])
    except:
        l.append(np.nan)
```

```python
img_url.extend(l)
```
